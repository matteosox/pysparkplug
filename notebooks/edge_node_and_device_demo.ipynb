```python
import datetime
import time

import pysparkplug as psp

host = "emqx"
group_id = "my_group"
edge_node_id = "my_edge_node"
device_id = "my_device"
metrics = (
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uint8",
        datatype=psp.DataType.UINT8,
        value=1,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uint16",
        datatype=psp.DataType.UINT16,
        value=2,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uint32",
        datatype=psp.DataType.UINT32,
        value=3,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uint64",
        datatype=psp.DataType.UINT64,
        value=4,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="int8",
        datatype=psp.DataType.INT8,
        value=-1,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="int16",
        datatype=psp.DataType.INT16,
        value=-2,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="int32",
        datatype=psp.DataType.INT32,
        value=-3,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="int64",
        datatype=psp.DataType.INT64,
        value=-4,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="float",
        datatype=psp.DataType.FLOAT,
        value=1.1,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="double",
        datatype=psp.DataType.DOUBLE,
        value=2.2,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="boolean",
        datatype=psp.DataType.BOOLEAN,
        value=True,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="string",
        datatype=psp.DataType.STRING,
        value="hello world",
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="datetime",
        datatype=psp.DataType.DATETIME,
        value=datetime.datetime(1990, 9, 3, 5, 4, 3),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="text",
        datatype=psp.DataType.TEXT,
        value="iamatext",
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uuid",
        datatype=psp.DataType.UUID,
        value="iamauuid",
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="bytes",
        datatype=psp.DataType.BYTES,
        value=b"iamabytes",
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="file",
        datatype=psp.DataType.FILE,
        value=b"iamafile",
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="int8_array",
        datatype=psp.DataType.INT8_ARRAY,
        value=(1, -2, 3),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="int16_array",
        datatype=psp.DataType.INT16_ARRAY,
        value=(4, -5, 6),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="int32_array",
        datatype=psp.DataType.INT32_ARRAY,
        value=(7, -8, 9),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="int64_array",
        datatype=psp.DataType.INT64_ARRAY,
        value=(10, -11, 12),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uint8_array",
        datatype=psp.DataType.UINT8_ARRAY,
        value=(1, 2, 3),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uint16_array",
        datatype=psp.DataType.UINT16_ARRAY,
        value=(4, 5, 6),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uint32_array",
        datatype=psp.DataType.UINT32_ARRAY,
        value=(7, 8, 9),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="uint64_array",
        datatype=psp.DataType.UINT64_ARRAY,
        value=(10, 11, 12),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="float_array",
        datatype=psp.DataType.FLOAT_ARRAY,
        value=(1.1, -2.2, 3.3),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="double_array",
        datatype=psp.DataType.DOUBLE_ARRAY,
        value=(4.4, -5.5, 6.6),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="boolean_array",
        datatype=psp.DataType.BOOLEAN_ARRAY,
        value=(True, False, True),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="string_array",
        datatype=psp.DataType.STRING_ARRAY,
        value=("hello", "world"),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="datetime_array",
        datatype=psp.DataType.DATETIME_ARRAY,
        value=(datetime.datetime(2024, 1, 1), datetime.datetime.now()),
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="null_uint8",
        datatype=psp.DataType.UINT8,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="historical_uint8",
        datatype=psp.DataType.UINT8,
        value=1,
        is_historical=True,
    ),
    psp.Metric(
        timestamp=psp.get_current_timestamp(),
        name="transient_uint8",
        datatype=psp.DataType.UINT8,
        value=1,
        is_transient=True,
    ),
)

edge_node = psp.EdgeNode(group_id, edge_node_id, metrics)
device = psp.Device(device_id, metrics)
edge_node.register(device)

edge_node.connect(host)
time.sleep(1)
edge_node.update(metrics)
time.sleep(1)
edge_node.update_device(device_id, metrics)
time.sleep(1)
edge_node.deregister(device_id)
time.sleep(1)
edge_node.disconnect()
```
